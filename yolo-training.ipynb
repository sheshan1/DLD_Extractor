# Validation

```python
# Load your trained model
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

# Validate with training settings
metrics = model.val(
    data="/kaggle/working/My2-1/data.yaml",
    imgsz=640,
    batch=16,
    conf=0.001,  # Default confidence threshold
    iou=0.6,     # Default IoU threshold
    plots=True    # Generate evaluation plots
)
```

```python
print(f"mAP50-95: {metrics.box.map}")     # Overall strictest metric
print(f"mAP50: {metrics.box.map50}")      # Primary benchmark
print(f"Precision: {metrics.box.p}")      # Low false positives
print(f"Recall: {metrics.box.r}")         # Few missed detections
```
